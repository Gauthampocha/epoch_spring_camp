```python
import numpy as np
#importing libraries
```

```python
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]
#data
```

# Data Preprocessing

```python
X=np.array([row[:3] for row in data])
y=np.array([row[3] for row in data])
a=np.array([1,0,0])
b=np.array([0,1,0])
c=np.array([0,0,1])
Y = [a if label == 'Apple' else b if label == 'Banana' else c for label in y]
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
y=np.array(Y)
y
```

```python
def euclidean_dist(x,y):
  return np.sum((x-y)**2)**0.5
```

```python
class KNN:
    def __init__(self,k=3):
      self.k=k

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)
    def predict_one(self,x):
      # finding distances to all points
        distances = [euclidean_dist(x, x_train) for x_train in self.X_train]
        #finding indices in a sorted manner
        distances_index=np.argsort(distances)
        #storing top k indices
        k_indices=distances_index[:self.k]
        #storing values of labels of k-nearest
        k_nearest=[self.y_train[i] for i in k_indices]
        #finding majority among k_nearest (using sum since it uses structure of one-hot encoding nicely)
        most_common=np.sum(k_nearest,axis=0)

        most_common=np.argmax(most_common)
        if most_common==0:
          return 'Apple'
        elif most_common==1:
          return 'Banana'
        else:
          return 'Orange'

    def predict(self,X_test):
        predictions=[self.predict_one(x) for x in X_test]
        return predictions



```

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])
test_data=(test_data-np.mean(test_data,axis=0))/np.std(test_data,axis=0)
knn1=KNN()
knn1.fit(X,y)
predictions1=knn1.predict(test_data)
print(predictions1)
#k=1,3 give expected output
#k=5 gives banana apple apple
#k=10 gives all 3 apple
```
